# Classification

### Importation des pakcages

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

### Création du dataset

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemple et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
y
```

# Graphiques de classification avec plt.scatter()

```python
plt.scatter(x[:,0], x[:,1], c = y, alpha = 0.5, s = x[:,2]*100) # y étant égale au 3 classes différentes
                                                                # alpha = transparence et s = taille des points
plt.xlabel('longueur sépal') # ici on ne peu représenter que 2 variables à la fois
plt.ylabel('largeur sépal')
```

# Graphiques 3D
### Importation d'un nouveau package

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection = '3d') # on crée un objet ax sur lequel on va pouvoir travailler
                                 # lorsqu'on crée des graphiques 3d avec matplotlib, on est obligé d'utiliser la POO

ax.scatter(x[:,0], x[:,1], x[:,2], c = y)
```

### Visualiser des surfaces et fonctions mathématiques

```python
f = lambda x,y:np.sin(x) + np.cos(x+y)
```

```python
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X, Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
Z.shape
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z, cmap = 'plasma')
```

# Histogrammes

### Visualisation de la distribution de nos données

```python
plt.hist(x[:,0], bins = 20) # bins = nb section dans notre histogramme
plt.hist(x[:,1], bins = 20)
```

### Histogramme 2D

```python
plt.hist2d(x[:,0],x[:,1], bins = 20, cmap = 'plasma')
plt.xlabel('largeur sépal')
plt.ylabel('longueur sépal')
plt.colorbar()
```

### Analyse d'une image avec un histogramme

```python
from scipy import misc
face = misc.face(gray=True)

plt.subplot(2,1,1)
plt.imshow(face, cmap ='gray')

plt.subplot(2,1,2)
plt.hist(face.ravel(), bins = 255) # fréquence d'apparition des difféente couleurs de pixel

plt.show()
```

# Graphiques contourplot
### - Chercher à résoudre des problèmes d'optimisation

```python
plt.contour(X, Y, Z, 20, colors = 'black') # 20 -> nb niveaux
```

```python
ff = lambda x,y: np.sin(x) + np.cos(x+y)*np.cos(x) # fonction un peu plus complexe pour imager

X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X, Y = np.meshgrid(X,Y)

Z = ff(X,Y)

plt.contour(X, Y, Z, 20, colors = 'black')
```

```python
plt.contourf(X, Y, Z, cmap = 'RdGy')
plt.colorbar()
```
