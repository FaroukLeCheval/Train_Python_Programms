# Le commencement
### - Toujours crée une figure en començant par plt.figure()
### - Dans le cycle de vie d'une figure on peu continuer a rajouter d'autre courbe sur notre figure

```python
plt.figure(figsize =(5,5)) #permet de choisir la taille de notre figure
plt.plot(x,y, label = 'linear')
plt.plot(x, x**2, label = 'squared')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('.../Results_prog/figure_1.png') # si il est mis après plt.show() il sauvegarde sur une nouveau plot vide
plt.show() # toujours à mettre même is dans jupyter on peu s'en passer
```

### Afficher plusieurs graphiques
##### fonction .subplot()

```python
plt.subplot(2, 1, 1) # On crée une grille avec subplot()
plt.plot(x,y**2, c='red')
plt.subplot(2, 1, 2)
plt.plot(x**2,y,c='red')
```

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot(x**2-6*x+1,y**2+19*y+1, label = 'croissance haute')
plt.plot(x**2+12*x+2,y**2-7*y+2, label = 'croissance longue')
plt.title('figure 1')
plt.legend()

plt.subplot(2,2,2)
plt.plot(x, np.sin(x),label='sinus')
plt.plot(x, np.cos(x),label='cosinus')
plt.title('figure 2')
plt.legend()

plt.show()
```

# Méthode POO

```python
fig, ax = plt.subplots(2, 1 , sharex=True) # le ax est un array numpy
ax[0].plot(x,y)
ax[1].plot(x**2,y/(1/y))
plt.show
```

# Exercice
##### - Créer une fonction graphique permettant de tracer sur une seule et même figure() , plusieurs graphique issue de données récolté dans plusieurs experiences. 
#####  - A la fin on obtient 4 graphique.

```python
dataset = {f"experience {i}": np.random.randn(100) for i in range (4)} # création d'un dataset
dataset
```

```python
print(dataset.keys())
```

```python
for cle, valeurs in dataset.items():
    print(f"{cle}")
    print(f"{valeurs[0]}")
```

```python
def graphique(dataset):
    nb_graphique = 1
    for cle, valeurs in dataset.items():
        plt.figure()
        plt.subplot(4,1,4)
        plt.plot(valeurs**3-((78*valeurs**4)/valeurs)*(1/valeurs)+100, label = 'neg')
        plt.plot(((valeurs**3)/(1/valeurs**3)-valeurs**2), label = 'pos')
        plt.title(f'figure {nb_graphique}')
        plt.legend()    
        nb_graphique+=1
        
graphique(dataset)
```
