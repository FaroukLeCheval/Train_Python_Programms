# Méthode .plot()

```python
x = range(1,100) # toujours avoir les même dimensions dans les tableaux x et y 
y = range(1,100)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,5,10)
y = np.linspace(0,5,10)
plt.plot(x,y) # plot(x,y, label = ...nom du graphique, lw = ...épaisseur du trait, ls = ...style du trait, c = ...couleur du trait)
plt.show()
```

# Le commencement
### - Toujours crée une figure en començant par plt.figure()
### - Dans le cycle de vie d'une figure on peu continuer a rajouter d'autre courbe sur notre figure

```python
plt.figure(figsize =(5,5)) #permet de choisir la taille de notre figure
plt.plot(x,y, label = 'linear')
plt.plot(x, x**2, label = 'squared')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('.../Results_prog/figure_1.png') # si il est mis après plt.show() il sauvegarde sur une nouveau plot vide
plt.show() # toujours à mettre même is dans jupyter on peu s'en passer
```
